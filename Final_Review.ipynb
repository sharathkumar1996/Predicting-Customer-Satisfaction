```python
orders = pd.merge(items, orders, on ="order_id")
orders.head()
```

```python
orders['order_estimated_days'] =orders['order_estimated_delivery_date'] - orders['order_approved_at']
orders['order_estimated_days'].head(5)
```

```python
orders['order_deliver_days'] = orders['order_delivered_customer_date'] - orders['order_approved_at']
orders['order_deliver_days'].head()
```

```python
orders['order_time_delta'] = orders['order_estimated_days'] - orders['order_deliver_days']
orders['order_time_delta'].head()
#if the number is poitive the item delivered before the estimated date
```

```python
 orders['Is_Late'] = orders.order_delivered_customer_date > orders.order_estimated_delivery_date
 orders['Is_Late'].head()
```

```python
orders['total_order_value'] = orders.price + orders.freight_value 

orders['order_freight_ratio'] = orders.price/orders.freight_value 

orders['average_product_value'] = orders.price / orders.order_item_id
```
